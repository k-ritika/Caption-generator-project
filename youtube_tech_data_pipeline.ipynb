```python
import pandas as pd
import re
import nltk
nltk.download('punkt')
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv("youtube_data.csv")
```

```python
def clean_text(text):
    text = re.sub(r"http\S+|www\S+|https\S+", '', text)
    text = re.sub(r'[^\w\s]', '', text)
    return text.lower()
```

**Exploring the Structure**

**Cleaning Data **

converting object(string) type data into datetime

```python
df['published_at'] = pd.to_datetime(df['published_at'], errors='coerce')
df['fetched_at'] = pd.to_datetime(df['fetched_at'], errors='coerce')
```

fill missing values :description since it is completely empty,synthesize it using title, keyword_tracked, and niche

```python
df['description']=(df['title'].astype(str) + " " + df['keyword_tracked'].astype(str) + " " + df['niche'].astype(str)).apply(clean_text)
```

```python
df['channel_name']=df['channel_name'].fillna("Unknown")
```

Feature Engineering
number of characters in description
engagement_rates=(comments+likes)/views
day_of_week=Monday.Tuesday,....

```python
df['desc_length']=df['description'].apply(len)
df['title_length']=df['title'].apply(len)
df['engagement_rate']=(df['comments']+df['likes'])/df['views']
df['day_of_week']=df['published_at'].dt.day_name()
df['month']=df['published_at'].dt.month_name()
df['hour']=df['published_at'].dt.hour
```

```python
df.index=range(1,len(df)+1)
df.head()
```
